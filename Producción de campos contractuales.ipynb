## Producción de petróleo y gas de 1985-2018

-Creacion de gráficas de la producción de petróleo y gas de los campos Artesa, Bacal, 5 presidentes, Giraldas, Lacamango de 1985-2018 y Juspi de 2013-2018.
-Los datos fueron tomdos de la Comisión Naional de Hidrocarburos (CNIH).



**Biblioteca a utilizar**
