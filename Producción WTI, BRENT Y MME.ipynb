# Precio del Petr칩leo WTI, BRENT Y MME 2005-2017

**Gr치fica Mensual Precio del Petr칩leo WTI, BRENT Y Mezcla Mexicana (MME) 2005-2017. Los datos fueron tomados de la p치gina Web del [Servico Geologico Mexicano (SGM)](http://www.sgm.gob.mx/Web/SINEM/energeticos/wti_brent_mme.html).**


**Importar y agrupar datos**

```python
% matplotlib inline 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

prod = pd.read_csv('Precio Petroleo-2005-2017.csv')
prod.columns
print ('tipo', type(prod))
WTI = prod ['WTI']
BRENT = prod['BRENT']
MME = prod['MME']
print ('WTI', type(WTI), 'BRENT', type(BRENT), 'MME', type('MME'))
```

**Grafica: Promedio Mensual 2005-2017**

```python
plt.figure(figsize=[15,8])
plt.style.use('bmh')

plt.plot(WTI, '-' , color = 'red')
plt.plot(BRENT, '-' , color = 'blue' )
plt.plot(MME, '-' , color = 'green' )
plt.grid(color='k', alpha=0.3,linestyle='--')

plt.xlabel('A침os', color='k', fontsize='x-large')

plt.ylabel('USD/Bbl', fontsize='x-large', color='k')

plt.title('Promedio Mensual Precio del Petr칩leo WTI,BRENT y Mezcla Mexicana', fontsize='x-large', color='black')
plt.xticks( [0,12,24,36,48,60,72,84,96,108,120,132,144],
           ['2005','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017'], 
               rotation=60, color='k')
plt.yticks( [0,20,40,60,80,100,120,140],
           ['0', '20', '40', '60', '80' , '100', '120', '140',], rotation=20, color='k')
plt.ylim(0,140)
plt.xlim(0,144)
plt.legend()
plt.show()
```
